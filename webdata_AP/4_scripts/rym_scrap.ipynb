```python
#importando as bibliotecas pandas e selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
#abrindo o broswer pelo webdriver
browser = webdriver.Firefox()
```

```python
#acessando o website que será utilizado no web scrap
browser.get('https://rateyourmusic.com/')
```

```python

browser.find_element(By.XPATH, '/html/body/header/div/div[3]/div/div[2]/a').click()
```

```python
browser.find_element(By.XPATH, '/html/body/div[6]/div/div[2]/sections/group[1]/section[1]/div/a[1]').click()
```

```python
#utilizo essa variável para encontrar todos os elementos relacionado a esse selector
all_album_name_elements = browser.find_elements(By.CSS_SELECTOR, '.page_section_charts_item_wrapper')

#então uso essa variável para iterar todos os elementos, podendo, assim, deixar de usar um valor estático
#nesse caso, utilizei o XPATH para encontrar o elemento
album_name_list = []
for element in all_album_name_elements:
    album_name = element.find_element(By.XPATH, './div/div[2]/div[1]/div[1]/div[1]/a/span/span').text
    album_name_list.append(album_name)

album_name_list
```

```python
all_artist_name_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_credited_links_primary')

#daqui em diante, apenas iterei os itens da lista, sem buscar pelo XPATH, já que os itens já estão na lista
artist_name_list = []
for element in all_artist_name_elements:
    artist_name = element.text
    artist_name_list.append(artist_name)
    
#a mesma lógica será usada em todos os outros loops (exceto os de album_name)
    
artist_name_list
```

```python
all_date_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_date')

#por algum motivo, os valores inseridos na lista de data estavam entrando duas vezes, para resolver isso, criei três listas e as concatenei
release_date_list_1 = []
for element in all_date_elements:
    release_date = element.text
    release_date_list_1.append(release_date)

```

```python
all_avg_rating_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_details_average_num')

avg_rating_list = []
for element in all_avg_rating_elements:
    avg_rating = element.text
    avg_rating_list.append(avg_rating)
```

```python
all_genre_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_genres_primary')

genre_list = []
for element in all_genre_elements:
    genre = element.text
    genre_list.append(genre)
```

```python
#passar para a segunda página e esperar 10 segundos para carregar todos os elementos
browser.find_element(By.XPATH, '/html/body/div[6]/div/div[2]/sections/group[2]/section[1]/div[2]/div[2]/div/span/a[1]').click()
browser.implicitly_wait(10)
```

```python
all_album_name_elements = browser.find_elements(By.CSS_SELECTOR, '.page_section_charts_item_wrapper')

for element in all_album_name_elements:
    album_name = element.find_element(By.XPATH, './div/div[2]/div[1]/div[1]/div[1]/a/span/span').text
    album_name_list.append(album_name)
```

```python
all_artist_name_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_credited_links_primary')

for element in all_artist_name_elements:
    artist_name = element.text
    artist_name_list.append(artist_name)
```

```python
all_date_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_date')

release_date_list_2 = []
for element in all_date_elements:
    release_date = element.text
    release_date_list_2.append(release_date)
    
release_date_list_2
```

```python
all_avg_rating_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_details_average_num')

for element in all_avg_rating_elements:
    avg_rating = element.text
    avg_rating_list.append(avg_rating)
```

```python
all_genre_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_genres_primary')

for element in all_genre_elements:
    genre = element.text
    genre_list.append(genre)
```

```python
browser.find_element(By.XPATH, '/html/body/div[6]/div/div[2]/sections/group[2]/section[1]/div[2]/div[2]/div/span/a[2]').click()
browser.implicitly_wait(10)
```

```python
all_album_name_elements = browser.find_elements(By.CSS_SELECTOR, '.page_section_charts_item_wrapper')

for element in all_album_name_elements:
    album_name = element.find_element(By.XPATH, './div/div[2]/div[1]/div[1]/div[1]/a/span/span').text
    album_name_list.append(album_name)
```

```python
all_artist_name_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_credited_links_primary')

for element in all_artist_name_elements:
    artist_name = element.text
    artist_name_list.append(artist_name)
```

```python
all_date_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_date')

release_date_list_3 = []
for element in all_date_elements:
    release_date = element.text
    release_date_list_3.append(release_date)
```

```python
all_avg_rating_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_details_average_num')

for element in all_avg_rating_elements:
    avg_rating = element.text
    avg_rating_list.append(avg_rating)
```

```python
all_genre_elements = browser.find_elements(By.CSS_SELECTOR, '.page_charts_section_charts_item_genres_primary')

for element in all_genre_elements:
    genre = element.text
    genre_list.append(genre)
```

```python
#list comprehension para eliminação de dados nulos que foram recebidos ao obter a variável rating 
#(para cada valor de rating existia um valor nulo)
filtered_avg_ratings = [rating for rating in avg_rating_list if rating]
filtered_avg_ratings
```

```python
#concatenação das listas de release_date
final_release_date_list = release_date_list_1 + release_date_list_2 + release_date_list_3
```

```python
#criação do Data Frame top_100_albuns_rym
top_100_albuns_rym = pd.DataFrame(album_name_list, columns=['Album Name'])
top_100_albuns_rym
```

```python
#inserção das colunas no Data Frame
top_100_albuns_rym['Artist'] = artist_name_list
top_100_albuns_rym['Release Date'] = final_release_date_list
top_100_albuns_rym['Average Rating'] = filtered_avg_ratings
top_100_albuns_rym['Genre'] = genre_list
```
